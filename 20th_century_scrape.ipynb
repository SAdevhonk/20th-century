```python
# Import required libraries
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# Define path to ChromeDriver
driver_path = "drivers/chromedriver-mac-arm64/chromedriver"

# Configure browser options
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Launch browser
driver = webdriver.Chrome(executable_path=driver_path, options=options)

# Open Wikipedia page
url = "https://en.wikipedia.org/wiki/Key_events_of_the_20th_century"
driver.get(url)

# Extract full page text
page_text = driver.find_element_by_tag_name("body").text

# Close browser
driver.quit()

# Save text to file
file_name = "20th_century_key_events.txt"
with open(file_name, "w", encoding="utf-8") as file:
    file.write(page_text)

print("Scraping completed successfully.")
```
